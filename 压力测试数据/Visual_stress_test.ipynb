# elasticsearch线程测试

## 吞吐量

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

current_directory = os.getcwd()

file_path = os.path.join(current_directory, 'pressureTest_result.xlsx')

data = pd.read_excel(file_path, sheet_name='elasticsearch线程测试')

samples = data['# Samples']
throughput = data['Throughput']

# 创建折线图
plt.figure(figsize=(10, 6))
plt.plot(samples, throughput, marker='o', linestyle='-', color='b')
plt.title('Elasticsearch Thread Testing')
plt.xlabel('Samples')
plt.ylabel('Throughput (requests per second - rps)')
plt.grid(True)
plt.show()
```
